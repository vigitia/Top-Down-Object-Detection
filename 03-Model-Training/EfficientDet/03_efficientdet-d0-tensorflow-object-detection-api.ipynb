### **Model configuration**
We downloaded and extracted a pre-trained model of our choice. Now we want to configure it.

**Model configuration** is a process that lets us tailor model-related artifacts (e.g. hyperparameters, loss function, etc) so that it can be trained (fine-tuned) to tackle detection for the objects that were interested in. Thats it.

The TensorFlow Object Detection API allows model configuration via the **pipeline.config** file that goes along with the pre-trained model.

```python
# Define the training pipeline
TRAINING_STEPS = 3000
WARMUP_STEPS = 200
NUM_CLASSES = 3
TRAIN_BATCH_SIZE = 4
FINE_TUNE_CHECKPOINT = "efficientdet_d0_coco17_tpu-32/checkpoint/ckpt-0"
LEARNING_RATE_BASE = 0.001
PIPELINE_CONFIG_PATH='data/pipeline.config'

pipeline = Template(config_file_template).substitute(
    training_steps=TRAINING_STEPS, 
    warmup_steps=WARMUP_STEPS, 
    learning_rate_base=LEARNING_RATE_BASE, 
    num_classes=NUM_CLASSES, 
    fine_tune_checkpoint = FINE_TUNE_CHECKPOINT,
    train_batch_size = TRAIN_BATCH_SIZE

)

with open(PIPELINE_CONFIG_PATH, 'w') as f:
    f.write(pipeline)
```

# Run inference on test images

```python
# Load the TensorFlow COTS detection model into memory.
start_time = time.time()
tf.keras.backend.clear_session()

# Load initial model
detect_fn_tf_odt = tf.saved_model.load(os.path.join('efficientdet_d0_coco17_tpu-32', 'saved_model'))

# Load finetuned model
#detect_fn_tf_odt = tf.saved_model.load(os.path.join(MODEL_DIR, 'output', 'saved_model'))
end_time = time.time()
elapsed_time = end_time - start_time
print('Elapsed time: ' + str(elapsed_time) + 's')
```

```python
import cv2

# Define some utils method for prediction.

def load_image(path):
    byte_img = cv2.imread(path)
    #byte_img = cv2.resize(byte_img, (640, 640))
    rgb = cv2.cvtColor(byte_img, cv2.COLOR_BGR2RGB)
    
    return tf.convert_to_tensor(rgb, dtype=tf.uint8)


def detect(image_np):
    input_tensor = np.expand_dims(image_np, 0)
    start_time = time.time()
    detections = detect_fn_tf_odt(input_tensor)
    return detections
```

```python
DETECTION_THRESHOLD = 0.3
#labels = {"1": "bowl", "2": "cup", "3": "bottle"}
labels = {"1": "person", "2": "bicycle", "3": "car", "4": "motorcycle", "5": "airplane", "6": "bus", "7": "train", "8": "truck", "9": "boat", "10": "traffic light", "11": "fire hydrant", "13": "stop sign", "14": "parking meter", "15": "bench", "16": "bird", "17": "cat", "18": "dog", "19": "horse", "20": "sheep", "21": "cow", "22": "elephant", "23": "bear", "24": "zebra", "25": "giraffe", "27": "backpack", "28": "umbrella", "31": "handbag", "32": "tie", "33": "suitcase", "34": "frisbee", "35": "skis", "36": "snowboard", "37": "sports ball", "38": "kite", "39": "baseball bat", "40": "baseball glove", "41": "skateboard", "42": "surfboard", "43": "tennis racket", "44": "bottle", "46": "wine glass", "47": "cup", "48": "fork", "49": "knife", "50": "spoon", "51": "bowl", "52": "banana", "53": "apple", "54": "sandwich", "55": "orange", "56": "broccoli", "57": "carrot", "58": "hot dog", "59": "pizza", "60": "donut", "61": "cake", "62": "chair", "63": "couch", "64": "potted plant", "65": "bed", "67": "dining table", "70": "toilet", "72": "tv", "73": "laptop", "74": "mouse", "75": "remote", "76": "keyboard", "77": "cell phone", "78": "microwave", "79": "oven", "80": "toaster", "81": "sink", "82": "refrigerator", "84": "book", "85": "clock", "86": "vase", "87": "scissors", "88": "teddy bear", "89": "hair drier", "90": "toothbrush"}

for image_name in [file for file in os.listdir('images/test/') if re.match(r'.*\.(jpg|jpeg|png)', file)]:
    print(image_name)
    image = load_image(os.path.join('images', 'test', image_name))
    height, width, _ = image.shape

    # Run object detection using the TensorFlow model.
    detections = detect(image)

    boxes, scores, classes, num_detections  = detections['detection_boxes'], detections['detection_scores'], detections['detection_classes'], detections['num_detections']
    pred_labels = classes.numpy().astype('int')[0]
    pred_labels = [labels[str(i)] for i in pred_labels]
    pred_boxes = boxes.numpy()[0]
    pred_scores = scores.numpy()[0]

    # Parse the detection result and generate a prediction string.
    num_detections = detections['num_detections'][0].numpy().astype(np.int32)
    image_copy = np.array(image).copy()
    
    for score, (ymin, xmin, ymax, xmax), label in zip(pred_scores, pred_boxes, pred_labels):
        if score < DETECTION_THRESHOLD:
            continue
        
        if label not in ['bowl', 'bottle', 'cup']:
            pass
        

        print(f"Label: {label}")
        xmin = int(xmin * width)
        ymin = int(ymin * height)
        xmax = int(xmax * width)
        ymax = int(ymax * height)
        
        y_offset = ymin-15
        x_offset = xmin - 5

        image_copy = cv2.rectangle(image_copy, (xmin, ymin), (xmax,ymax), (0,255,0), 2)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(image_copy, label, (x_offset + 100, y_offset), font, 0.75, (255, 0, 0), 2, cv2.LINE_AA )
        score_txt = f"{round(score * 100, 2)}%"
        print(score_txt)
        cv2.putText(image_copy, score_txt, (x_offset, y_offset), font, 0.75, (255, 0, 0), 2, cv2.LINE_AA )
        break

    filename = os.path.join('efficientdet_d0_coco17_tpu-32', 'pred_test', image_name).split('.')[0]
    plt.imshow(image_copy)
    plt.savefig(f"{filename}.jpg", dpi=150)
    plt.show()
```
